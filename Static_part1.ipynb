```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
# Function to perform data analysis
def perform_data_analysis(data):
    # Automatically identify the target variable column
    target_column = None
    for column in data.columns:
        if column.lower() == 'target' or 'label' in column.lower():
            target_column = column
            break

    if target_column is None:
        print("No target variable found in the dataset. Please specify the correct column name.")
        return

    # Check the first few rows of the dataset
    print(data.head())

    # Check the distribution of the target variable
    sns.countplot(data['Target Attack'])
    plt.title("Target Attack Distribution")
    plt.show()

    # Check the distribution of each feature
    for feature in data.columns:
        if feature != target_column:
            plt.figure(figsize=(8, 5))
            sns.histplot(data[feature], kde=True)
            plt.title(f"Distribution of {feature}")
            plt.show()

    # Calculate and display summary statistics
    summary_stats = data.describe()
    print(summary_stats)

    # Check for data skewness
    skewness = data.skew()
    print("Skewness for each feature:")
    print(skewness)

    # Visualize correlation between features (if needed)
    plt.figure(figsize=(10, 8))
    sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
    plt.title("Feature Correlation Heatmap")
    plt.show()

# Perform data analysis
perform_data_analysis(data)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, chi2, f_classif
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Generate a synthetic dataset
np.random.seed(42)
 

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Feature Selection based on Correlation
correlation_scores = X_train.corrwith(y_train).abs().sort_values(ascending=False)
selected_features_correlation = correlation_scores.index[:5]  # Select the top 5 features

# Feature Selection using Chi-Square Test (for categorical features)
chi2_scores, _ = chi2(X_train, y_train)
selected_features_chi2 = X_train.columns[np.argsort(chi2_scores, kind="heapsort")[::-1]][:5]

# Train a model with the selected features
X_train_correlation = X_train[selected_features_correlation]
X_train_chi2 = X_train[selected_features_chi2]

model_correlation = RandomForestClassifier(random_state=42)
model_chi2 = RandomForestClassifier(random_state=42)

model_correlation.fit(X_train_correlation, y_train)
model_chi2.fit(X_train_chi2, y_train)

# Evaluate the models on the test set
X_test_correlation = X_test[selected_features_correlation]
X_test_chi2 = X_test[selected_features_chi2]

y_pred_correlation = model_correlation.predict(X_test_correlation)
y_pred_chi2 = model_chi2.predict(X_test_chi2)

accuracy_correlation = accuracy_score(y_test, y_pred_correlation)
accuracy_chi2 = accuracy_score(y_test, y_pred_chi2)

print(f"Accuracy with Correlation-Selected Features: {accuracy_correlation:.2f}")
print(f"Accuracy with Chi-Square-Selected Features: {accuracy_chi2:.2f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split

def evaluate_model_chi(model, x, y):
    y_pred = model.predict(x)
    f1 = f1_score(y, y_pred)
    return f1

# Load your dataset
data2 = pd.read_csv("static_dataset.csv")
data2 = data2.dropna()
data2 = data2.drop(columns=['timestamp'])

selected_columns = ['sld', 'longest_word', 'numeric', 'subdomain_length', 'FQDN_count',
                    'special', 'upper', 'labels']

categorical_cols = [col for col in data2.columns if data2[col].dtype == 'object']

# Convert categorical columns to numeric using LabelEncoder
for column in categorical_cols:
    label_encoder = LabelEncoder()
    data2[column] = label_encoder.fit_transform(data2[column])

y2 = data2['Target Attack']
x_noise = data2[selected_columns]

# Add noise to the data
noise = np.random.normal(0, 0.1, size=x_noise.shape)
x_noise += noise

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x_noise, y2, test_size=0.2, random_state=42)

# Define the range of values for n_estimators and max_depth
n_estimators_values = [100, 120, 150]
max_depth_values = [1, 2, 3]

# Lists to store results
f1_scores = []

# Loop through different combinations of n_estimators and max_depth
for n_estimators in n_estimators_values:
    for max_depth in max_depth_values:
        # Train the random forest model
        model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
        model.fit(X_train, y_train)

        # Evaluate the model
        f1 = evaluate_model_chi(model, X_test, y_test)
        f1_scores.append(f1)

        # Print model performance on the validation set
        print(f"Random Forest Metrics for n_estimators={n_estimators}, max_depth={max_depth}:")
        print(f"F1-Score: {f1:.2f}")
        print("=" * 20)

# Print the list of F1 scores for each combination
print("List of F1 Scores:")
print(f1_scores)
```
